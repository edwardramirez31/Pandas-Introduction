# SCATTER PLOTS
Gráficos usados para ver si dos conjuntos de datos están correlacionados, observar tendencias, valores atípicos, etc. Para graficar un scatter plot en Matplotlib se usa `plt.scatter`

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.style.use('seaborn')
# no hay correlación entre ambas listas de datos
plt.scatter(x, y)

plt.tight_layout
plt.show
```

# PERSONALIZACIÓN DEL GRÁFICO
- El argumento `s` usado para establecer el tamaño de los puntos
- El argumento `c` para establecer un color
- El argumento `marker` para el tipo de punto
- `edgecolor` para darle borde del círculo
- `linewidth` para darle el ancho al borde
- `Alpha` para la transparencia


```python
plt.scatter(x, y, marker='o', s=100, edgecolor='black', linewidth=1.5, alpha=0.5, c='blue')


plt.tight_layout()
plt.show()
```

```python
# pasar un color diferente para cada punto
# cada color representa la intensidad donde 0 tiende a blanco y 9 a ser más oscuro en la paleta de colores blanco y negro

# los colores significan la satisfacción del usuario con los productos
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, c=colors, alpha=0.75, linewidth=1.5, s=100, edgecolor='black')


plt.xlabel('Cantidad de Anuncios')
plt.ylabel('Cantidad de artículos comprados')
plt.title('Relacion entre Articulos Comprados por Anuncio Por Mes')
plt.show
```

# MÁS PALETAS DE COLORES
Para cambiar la paleta de colores, se pueden usar los [colormaps](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html) que provee Matplotlib a través del argumento `cmap`

```python
# cmap argument
plt.scatter(x, y, c=colors, alpha=0.75, linewidth=1.5, s=100, edgecolor='black', cmap='hot')


plt.xlabel('Cantidad de Anuncios')
plt.ylabel('Cantidad de artículos comprados')
plt.title('Relacion entre Articulos Comprados por Anuncio Por Mes')
plt.show
```

Para crear una leyenda de colores, se usa el método `plt.colorbar` y se le asigna una etiqueta con el método `set_label`

```python
# labels a los colores
# color bar legend
plt.scatter(x, y, c=colors, cmap='Reds', edgecolor='black', linewidth=1.5, s=100, alpha=0.8)

cbar = plt.colorbar()
cbar.set_label('Satisfacción')

plt.xlabel('Cantidad de Anuncios')
plt.ylabel('Cantidad de artículos comprados')
plt.title('Relacion entre Articulos Comprados por Anuncio Por Mes')

plt.show
```

# Plotear los tamaños de los puntos
Algunos scatter plots representan los puntos de diferentes tamaños, indicando así información adicional y útil para el gráfico. Esto se puede hacer pasando una lista donde cada elemento representa el tamaño de su respectivo punto

```python
# tamaño que corresponde a cada punto del scatter plot
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
plt.scatter(x, y, s=sizes, c=colors, alpha=0.8, cmap='Blues', edgecolor='black', linewidth=1.5)

c = plt.colorbar()
c.set_label('Satisfacción')

plt.xlabel('Cantidad de Anuncios')
plt.ylabel('Cantidad de artículos comprados')
plt.title('Relacion entre Articulos Comprados por Anuncio Por Mes')

plt.tight_layout()
plt.show
```

# Youtube Likes and Views Example
En el archivo `youtube_data.csv` se encuentran diversos datos que representan la cantidad de vistas, likes y el ratio de like/dislike. Se usa un Scatter Plot para observar la relación entre las vistas y los likes.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv('youtube_data.csv')
views = dataframe['view_count']
likes = dataframe['likes']
ratio = dataframe['ratio']
dataframe
```

```python
[dataframe.loc[i, 'view_count'] / dataframe.loc[i, 'likes']  for i in range(5)]
```

```python
plt.scatter(views, likes, c=ratio, edgecolor='black', alpha=0.8, linewidth=1.5, cmap='Blues')

color = plt.colorbar()
color.set_label('Ratio')

plt.xlabel('Views')
plt.ylabel('Likes')
plt.title('Views and Likes Relationship')

plt.tight_layout()

plt.show()
```

```python
plt.scatter(views, likes, c=ratio, edgecolor='black', alpha=0.8, linewidth=1.5, cmap='summer')

color = plt.colorbar()
color.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Total Views')
plt.ylabel('Total Likes')
plt.title('Views and Likes Relationship')

plt.tight_layout()

plt.show()
```
