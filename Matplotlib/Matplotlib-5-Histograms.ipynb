# HISTOGRAMAS:
Usados para visualizar la distribucion de datos que caen dentro de ciertos límites. El histograma agrupa los datos dentro de cierto rango o contenedor en el `eje x` en lugar de plotear cada valor de forma individual como sucede con los gráficos de barras.
> Yo mismo puedo indicarle a Matplotlib cuál es el rango o contenedor de datos y este se encarga de plotear la cantidad de datos pertenecientes a ese rango

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

plt.style.use('seaborn-notebook')
# En este caso, Matplotlib asignará el rango 
plt.hist(ages, edgecolor='black')
plt.title('Age of Respondents')
plt.xlabel("Ages")
plt.ylabel("Total Responses")

plt.show
```

Para indicar los rangos donde se agrupan los datos se pueden usar dos opciones:
- Pasar un integer el cuál indica la cantidad de tramos o rangos 
- Pasar una lista con el rango en el argumento `bins`

```python
plt.hist(ages, bins=5, edgecolor='black')


plt.title('Age of Respondents')
plt.xlabel("Ages")
plt.ylabel("Total Responses")

plt.show
```

```python
bins=[10, 20, 30, 40, 50, 60]
plt.hist(ages, bins=bins, edgecolor='black')


plt.title('Age of Respondents')
plt.xlabel("Ages")
plt.ylabel("Total Responses")

plt.show
```

```python
# Si quiero que no aparezca cierta parte del gráfico
# No coloco dicho rango dentro de bins
bins=[20, 30, 40, 50]
plt.hist(ages, bins=bins, edgecolor='black')


plt.title('Age of Respondents')
plt.xlabel("Ages")
plt.ylabel("Total Responses")

plt.show
```

Se procede a cargar datos de una encuesta donde cada fila corresponde a la edad escogida por el participante, con el fin de mostrar una distribución de las edades de aquella encuesta.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv('ages.csv', index_col='Responder_id')
dataframe
```

```python
ages = dataframe['Age']
```

```python
bins = [10, 20, 30, 40, 50, 60, 70, 80, 90]
plt.hist(ages, bins=bins, edgecolor='black')


plt.title('Age of Respondents')
plt.xlabel("Ages")
plt.ylabel("Total Responses")

plt.tight_layout()
plt.show
```

# ESCALA LOGARITMICA

En caso de que los datos de cierto rango sean tan grandes que opacan y no permiten visualizar la cantidad de otros rangos, se puede pasar el atributo `log=True`

```python
plt.hist(ages, bins=bins, edgecolor='black' log=True)

plt.title('Age of Respondents')
plt.xlabel("Ages")
plt.ylabel("Total Responses")

plt.tight_layout()
plt.show
```

Se pueden agregar más datos a esta gráfica, por ejemplo, un corte con el eje x usando el método `axvline` para mostrar cuál es la edad media de la encuesta

```python
avg_age = 29

plt.hist(ages, bins=bins, edgecolor='black')
plt.axvline(x=avg_age, color='red', label='Average Age', linewidth=2)

plt.legend()

plt.title('Age of Respondents')
plt.xlabel("Ages")
plt.ylabel("Total Responses")

plt.tight_layout()
plt.show
```

# Colocar líneas 

Otra forma de colocar líneas es usar los métodos `hlines` y `vlines` a los cuales se les  puede pasar:
- Una lista de cortes con los ejes o solo un corte
- El máximo y mínimo
- Colores
- Estilos de líneas
- Ancho de línea
- Label

> Para más ejemplos leer [Stack OverFlow](https://stackoverflow.com/questions/33382619/plot-a-horizontal-line-using-matplotlib)

```python
import numpy as np

xs = np.linspace(1, 21, 200)

plt.figure(figsize=(6, 3))
plt.hlines(y=39.5, xmin=100, xmax=175, colors='aqua', linestyles='-', lw=2, label='Single Short Line')
plt.hlines(y=[39, 40, 41], xmin=[0, 25, 50], xmax=[len(xs)], colors=['purple', 'blue', 'red'], 
           linestyles=['--', '-.', ':'], lw=3, label='Multiple Lines')
plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)

plt.show
```
