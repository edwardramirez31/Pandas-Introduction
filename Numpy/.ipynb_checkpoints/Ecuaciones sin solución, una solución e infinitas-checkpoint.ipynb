```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-6, 6)
# Tenemos más ecuaciones que variables: NO TIENE SOLUCIÓN
y_1 = 3*x + 5
y_2 = -x + 3
y_3 = 2*x + 1
```

```python
plt.style.use('seaborn-notebook')
plt.figure()

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

#limites
plt.xlim(-8, 8)
plt.ylim(-8, 8)

# Dibujar ejes
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
```

```python
# SISTEMA DE ECUACIONES QUE TIENE UNA SOLUCIÓN
# Solo pueden haber 0, 1 o infinitas soluciones
plt.plot(x, y_2)
plt.plot(x, y_3)

#limites
plt.xlim(-8, 8)
plt.ylim(-8, 8)

# Dibujar ejes
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
```
